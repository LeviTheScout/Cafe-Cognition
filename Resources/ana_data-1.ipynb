```python
# Filter data for 'Combo-1' and 'Combo-3' items
combo_df = df[df['Item Name'].isin(['Combo-1', 'Combo-3'])]

# Group data by Date, Item Name, and sum the quantities
grouped_df = combo_df.groupby(['Date', 'Item Name'])['Qty.'].sum().reset_index()

# Pivot the DataFrame to get it in the right format for a stacked area chart
df_pivot = grouped_df.pivot_table(index='Date', columns='Item Name', values='Qty.', aggfunc='sum', fill_value=0)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))
df_pivot.plot.area(ax=ax, stacked=True)

# Customize the plot
ax.set_title('Quantities Sold of Combo-1 and Combo-3 Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Quantity Sold')
plt.legend(title='Item Name', loc='upper left')
plt.show()
```

```python
def calssify(string):
    if string in ['Burger','Maggi', 'Frankie', 'Pizza', 'French Fry (180Gms)','Dabeli', 'Sandwich', 'Vadapav','Garlic Bread & Toast', 'Pasta','Slice','Puff', 'Vip Grill Sandwich','Vip Club Sandwich', 'Hard Base Pizza', 'Chaat']:
        return 'Food'
    elif string=='Combo':
        return 'Combo'
    else:
        return 'Beverage'
```

```python
df['Class']=df['Category'].apply(calssify)
```

```python
df.groupby('Class')['Qty.','Final Total'].sum()
```

```python
plt.pie(df.groupby('Class')['Final Total'].sum(),labels=['Beverage','Combo','Food'],autopct='%.1f%%')
plt.title('Revenu Distribution of ')
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Assuming 'Class' is a column in your DataFrame, and 'Final Total' is the revenue column

# Grouping by 'Class' and summing 'Final Total'
class_revenue = df.groupby('Class')['Final Total'].sum()

# Plotting the pie chart
fig, ax = plt.subplots()
ax.pie(class_revenue, labels=['Beverage', 'Combo', 'Food'], autopct='%.1f%%')

# Adding a title
plt.title('Revenue Distribution')

# Save the figure with a transparent background
plt.savefig('pie_chart.png', transparent=True)

# Display the plot
plt.show()
```

```python

combo1_filt = (df['Item Name'] == 'Combo-1')
combo1_counts = df.loc[combo1_filt].groupby(df['Date'].dt.month)['Final Total'].count()

combo3_filt = (df['Item Name'] == 'Combo-3')
combo3_counts = df.loc[combo3_filt].groupby(df['Date'].dt.month)['Final Total'].count()

counts_df = pd.DataFrame({'Combo-1': combo1_counts, 'Combo-3': combo3_counts}).fillna(0)

counts_df = counts_df[['Combo-1', 'Combo-3']]

fig, ax = plt.subplots(figsize=(10, 6))
counts_df.plot.area(ax=ax, stacked=True)

ax.set_title('Counts of Combo-1 and Combo-3 Sales Over Months')
ax.set_xlabel('Month')
ax.set_ylabel('Count')
plt.legend(title='Item Name', loc='upper left')
plt.show()
```
